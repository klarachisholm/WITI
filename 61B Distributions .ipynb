Plot grade distributions of CS10 in 2015, grouped by race.

```python
cs10_2015data = cs10_2015.merge(sp07_17, on='ppsk')
cs102015_race = cs10_2015data.groupby(['grade.name', 'ethnic.l1']).count()
cs102015_race = cs102015_race.reset_index()[['ppsk', 'grade.name', 'ethnic.l1']]
# Set order of grade names for plot
order = ["A+", "A", "A-", "B+", "B", "B-", "C+", "C", "C-", "D+", "D", "D-", "F", "Pass", "Incomplete", "Not Pass", "Withdrawn"]
# Set plot size
plt.figure(figsize=(40,20))
# Set title
# Plot
cs102015_raceplot = sns.barplot(x="grade.name", y="ppsk", hue="ethnic.l1", data=cs102015_race, 
                                        hue_order=['Asian/Pacific Islander', 'International', 'White/Other',
                                                   'Underrrepresented Minority'], order=order)
cs102015_raceplot.axes.set_title("CS10 2015 Grade Distribution (Counts by Race)",fontsize=50)
cs102015_raceplot.set_xlabel("Grade",fontsize=50)
cs102015_raceplot.set_ylabel("Counts of Students",fontsize=50)
cs102015_raceplot.tick_params(labelsize=30)
plt.legend(fontsize='x-large', title_fontsize='40')
plt.show()
```
